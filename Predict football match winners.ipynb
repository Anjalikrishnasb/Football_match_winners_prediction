```python
#convert object value to date time format
df['date']=pd.to_datetime(df['date'])
```

```python
#this is used to convert the data type to categorical datatype and ".cat.codes" that helps to assign unique numerical code to
#each unique category in the venue column.
df['venue_code']=df['venue'].astype("category").cat.codes
df.head()
```

```python
#Regular expression-regex
#While performing this code,that helps to remove everything after the first colon(:)
df['hour']=df['time'].str.replace(":.+","",regex=True).astype("int")
```

```python
df['target']=(df['result']=='W').astype("int")
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=50,min_samples_split=10,random_state=1)
```

```python
train=df[df['date']<'2022-01-01']
test=df[df['date']>'2022-01-01']
predictors=['venue_code','opponent_code','day_code']
```

```python
model.fit(train[predictors],train['target'])
```

```python
y_pred=model.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(test['target'],y_pred)
accuracy
```

```python
#This creates a python dictionary,where keys are actual and prediction,and the values are test['target'] and y_pred.
combined=pd.DataFrame(dict(actual=test['target'],prediction=y_pred))
```

```python
group=df.groupby('team')
```

```python
#rolling averages are also known as moving averages,this is used to analyze the data points by creating a series of averages of-
#different subset of the whole data set.
def rolling_averages(group,cols,new_cols):
    group=group.sort_values("date")
    rolling_stats=group[cols].rolling(3,closed='left').mean()#3 represent the window size
    group[new_cols]=rolling_stats
    group=group.dropna(subset=new_cols)
    return group
```

```python
df.columns
```

```python
cols=['gf','ga','sh','sot','dist','fk','pk','pkatt']
new_cols=[f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group,cols,new_cols)
```

```python
df_rolling=df.groupby('team').apply(lambda x:rolling_averages(x,cols,new_cols))
```

```python
#droplevel is a method used to removes one or more levels from multi index datadrame.
df_rolling=df_rolling.droplevel('team')
```

```python
def make_predictions(df,predictors):
    train=df[df['date']<'2022-01-01']
    test=df[df['date']>'2022-01-01']
    model.fit(train[predictors],train['target'])
    y_pred=model.predict(test[predictors])
    combined=pd.DataFrame(dict(actual=test['target'],prediction=y_pred))
    precision=precision_score(test['target'],y_pred)
    return combined,precision
```
